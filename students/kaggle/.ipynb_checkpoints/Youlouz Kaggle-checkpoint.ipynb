# Kmeans

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
train_set, test_set, train_sales, test_sales = train_test_split(train, train['SalePrice'], test_size = 0.25)
train_set.drop(['SalePrice'],axis=1,inplace=True);test_set.drop(['SalePrice'],axis=1,inplace=True)
X = train_set.select_dtypes(['int64','float64'])
Y = test_set.select_dtypes(['int64','float64'])
X_scaled = preprocessing.scale(X)
Y_scaled = preprocessing.scale(Y)
kmeans = KMeans(n_clusters=500, random_state=0).fit(X_scaled)

L=train_sales.groupby(kmeans.labels_).mean()
kmeans_train_pred=[L[kmeans.labels_[i]]for i in range(0,len(kmeans.labels_))]

#M=test_sales.groupby(kmeans.predict(Y_scaled)).mean()
kmeans_test_pred=[L[kmeans.predict(Y_scaled)[i]]for i in range(0,len(kmeans.predict(Y_scaled)))]
print sum((train_sales-kmeans_train_pred)**2)/sum((train_sales-train_sales.mean())**2)
print sum((test_sales-kmeans_test_pred)**2)/sum((test_sales-test_sales.mean())**2)
```

# ANOVA (effet des variables categorielles) + regression linéaire

```python
###supprimmer les na
if 'FireplaceQu' in train.columns:
    train=train.drop(['FireplaceQu'],axis=1).dropna()
#### variance de SalePrice par rapport aux variables catégorielles
t=[]
for j in range(0,len(train.select_dtypes("object").columns)):
    w=[]
    for i in range(0,len(train[train.select_dtypes("object").columns[j]].value_counts().index)):
        a= train['SalePrice'][train[train.select_dtypes("object").columns[j]]==train[train.select_dtypes("object").columns[j]].value_counts().index[i]].mean()
        w.append(a)
    t.append(np.sqrt(np.var(w))/train['SalePrice'].mean())
    
train.select_dtypes('object').columns[[x>=0.4 for x in t]]
```

```python
l=[list(train['Condition2'])[i]+list(train['ExterQual'])[i]+list(train['BsmtQual'])[i]+list(train['KitchenQual'])[i] for i in range(0,train.shape[0])]
train['Concatenate']=l
###la moyenne de SalePrice par catégorie
p=[]
for i in list(set(l)):
    p.append(train['SalePrice'][[list(train['Concatenate'])[j]==i for j in range(0,train.shape[0])]].mean())
### déduction de l'effet moyen de chaque catégorie à partir de la colonne SalePrice
effet_moy=[]
a=train['SalePrice'].mean()
for i in range(0,train.shape[0]):
    effet_moy.append(p[list(set(l)).index(list(train.Concatenate)[i])]-a)
train['NewSalePrice']=train['SalePrice']-effet_moy
#### regression linéaire de NewSalePrice par rapport aux variables numériques
from sklearn.model_selection import train_test_split
train_set, test_set, train_sales, test_sales = train_test_split(train, train['SalePrice'], test_size = 0.20)
X_train=train_set.select_dtypes(['int64','float64']).drop('SalePrice',axis=1)
X_test=test_set.select_dtypes(['int64','float64']).drop('SalePrice',axis=1)
Y_train=X_train['NewSalePrice']
Y_test=X_test['NewSalePrice']
X_train=X_train.drop('NewSalePrice',axis=1)
X_test=X_test.drop('NewSalePrice',axis=1)
regr = LinearRegression()
regr.fit(X_train, Y_train)
#####predictions
predictions_train=regr.predict(X_train)+[p[list(set(l)).index(list(train_set.Concatenate)[i])]-a for i in range (0,X_train.shape[0])]
predictions_test=regr.predict(X_test)+[p[list(set(l)).index(list(test_set.Concatenate)[i])]-a for i in range (0,X_test.shape[0])]
from sklearn.metrics import mean_squared_error, r2_score
#print('Variance score: %.2f' % r2_score(train_sales, predictions))
print sum((train_sales-predictions_train)**2)/sum((train_sales-train_sales.mean())**2)
print sum((test_sales-predictions_test)**2)/sum((test_sales-test_sales.mean())**2)
```

# GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
train_set, test_set, train_sales, test_sales = train_test_split(train, train['SalePrice'], test_size = 0.25)
train_set.drop(['SalePrice'],axis=1,inplace=True);test_set.drop(['SalePrice'],axis=1,inplace=True)

train_set_scaled = preprocessing.scale(train_set.select_dtypes(['int64','float64']))
test_set_scaled = preprocessing.scale(test_set.select_dtypes(['int64','float64']))

est = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,max_depth=5, random_state=0, loss='ls').fit(train_set_scaled,train_sales)
predictions_train=est.predict(train_set_scaled)
predictions_test=est.predict(test_set_scaled)
from sklearn.metrics import r2_score
print('Variance score pour train: %.2f' % r2_score(train_sales,predictions_train ))
print('Variance score pour test: %.2f' % r2_score(test_sales,predictions_test ))
est.feature_importances_
```

## Cross validation with Grid Research

```python
from sklearn.grid_search import GridSearchCV
parametres = {"learning_rate":[0.3,0.2,0.1,0.05,0.01],"max_depth":[2,3,4,5,6,10],"subsample":[1.0,0.8,0.5]}
gbr = GradientBoostingRegressor()
grille = GridSearchCV(estimator=gbr,param_grid=parametres,scoring="r2")
resultats = grille.fit(train_set_scaled,train_sales)
# best combination of parameters : {'subsample': 0.5, 'learning_rate': 0.2, 'max_depth': 4}
print(resultats.best_params_)
# prediction with the ‘’model’’ identified by cross-validation
ypredc = resultats.predict(test_set_scaled)
# performances of the ‘’best’’ model: test error rate = 0.065
err_best = resultats.best_score_
print(err_best)
```
