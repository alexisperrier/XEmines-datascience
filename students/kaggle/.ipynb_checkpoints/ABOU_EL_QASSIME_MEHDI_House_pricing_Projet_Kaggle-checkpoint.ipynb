```python
#Eleminer Skiewness des variables 
from scipy.stats import norm, skew
numeric_feats = adf.dtypes[adf.dtypes != "object"].index
skewed_feats = adf[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness = skewness[abs(skewness) > 0.75]
from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.15
for feat in skewed_features:
    adf[feat] = boxcox1p(adf[feat], lam)
```

# Preparation pour le modeling ! 

```python
df = adf[:n]
vdf = adf[n:]
X_valid = vdf
```

```python
x=df
```

```python
df.sample(frac = 1)
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.30, random_state=2)
```

```python
import numpy as np
import pandas as pd
import datetime

import random
import matplotlib.pyplot as plt

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.kernel_ridge import KernelRidge

from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet 
from sklearn.svm import SVR

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
import xgboost as xgb
from sklearn.model_selection import KFold, cross_val_score, train_test_split

from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
#K_folds et fonction de calcul de la metric 
n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(X_train)
    rmse= np.sqrt(-cross_val_score(model, X_test, Y_test, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

# Des mod√®les ... 

```python
model_xgb = xgb.XGBRegressor(colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=2200,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213, silent=1,
                             random_state =7, nthread = -1)
model_xgb.fit(X_train, Y_train)
xgb_train_pred = model_xgb.predict(X_train)
xgb_pred = np.expm1(model_xgb.predict(X_test))
```

```python
score = rmsle_cv(model_xgb)
print("xgb score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
GBoost = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =5)
GBoost.fit(X_train, Y_train)
xgb_train_pred = GBoost.predict(X_train)
GBoost_pred = np.expm1(GBoost.predict(X_test))
```

```python
score = rmsle_cv(GBoost)
print("GBoost score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
lasso = make_pipeline(RobustScaler(), Lasso(alpha =0.0005, random_state=1))
lasso.fit(X_train, Y_train)
xgb_train_pred = lasso.predict(X_train)
lasso_pred = np.expm1(lasso.predict(X_test))
score = rmsle_cv(lasso)
print("Lasso score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
KRR = KernelRidge(alpha=0.6, kernel='polynomial', degree=1, coef0=2.5)
KRR.fit(X_train, Y_train)
xgb_train_pred = KRR.predict(X_train)
KRR_pred = np.expm1(KRR.predict(X_test))
score = rmsle_cv(KRR)
print("KRR score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
ENet = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0005, l1_ratio=.9, random_state=3))
ENet.fit(X_train, Y_train)
xgb_train_pred = ENet.predict(X_train)
ENet_pred = np.expm1(ENet.predict(X_test))
score = rmsle_cv(ENet)
print("ENet score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
GBoost_pred = np.expm1(GBoost.predict(X_valid))
lasso_pred = np.expm1(lasso.predict(X_valid))
KRR_pred = np.expm1(KRR.predict(X_valid))
ENet_pred = np.expm1(ENet.predict(X_valid))
xgb_pred = np.expm1(model_xgb.predict(X_valid))
```

#### Apres plusieurs mod√®les, et plusieurs submissions, j'arrive a trouver mon dernier mod√®le de prediction, qui est une combinaison de plusieurs autres !! 

```python
Avereage_pred = xgb_pred*1/4 + GBoost_pred*1/4  + ENet_pred*1/4 + KRR_pred*1/4
```

```python
#Submission 
Y_valid = Avereage_pred
results = pd.DataFrame(columns = ['Id', 'SalePrice'])

# Kaggle veut que la colonne index commence par 1461
results['Id'] = X_valid.index + 1

results['SalePrice'] = Y_valid

# ecrire le resultats dans le fichier csv
results.to_csv("submission_1.csv", index = False)
```
