```python
from sklearn.datasets import make_regression
from time import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as se
%matplotlib inline 
```

```python
df = pd.read_csv('advertising.csv')
df.head()
```

```python
df.shape
```

```python
# The distribution of each variable
for i in range(len(df.iloc[1,:])):
        figure = plt.figure()
        plt.hist(df.iloc[:,i], bins=30)
        plt.title('Histogram_Variable %s' %df.columns[i])
```

```python
# Visualization of outliers
for i in range(len(df.iloc[1,:])):
        figure = plt.figure()
        plt.boxplot(df.iloc[:,i])
        plt.title('Boxplot_Variable %s' %df.columns[i])
#Outliers in Newspaper
```

```python
# Remove Outliers 
condition = (df['Newspaper'] < 85)
df=df[condition]
df.shape
```

```python
df.isna().sum()
#No missing values
```

```python
# Correlation between variables and correlation matrix (heatmap)
corr = df.corr()
se.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
```

```python
#Scatterplot of each 2 variables
for i in range (-1,3):
    figure = plt.figure()
    plt.scatter(df.iloc[:,i], df.iloc[:,i+1])
    plt.xlabel('{}'.format(df.columns[i]))
    plt.ylabel('{}'.format(df.columns[i+1]))
```

```python
#Scatterplot of each 2 variables
for i in range (-1,1):
    figure = plt.figure()
    plt.scatter(df.iloc[:,i], df.iloc[:,i+2])
    plt.xlabel('{}'.format(df.columns[i]))
    plt.ylabel('{}'.format(df.columns[i+2]))
 # Significant correlation between Radio and sales , TV and sales
```

```python
# univariate linear regression(Sales = f(Radio))
import statsmodels.formula.api as smf
lm_Radio = smf.ols(formula='Sales ~ Radio ', data=df).fit()
lm_Radio.pvalues
# p value = 9*e-20
```

```python
lm_Radio.rsquared
# R-squared(Radio) = 0.35
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(df['Sales'], lm_Radio.fittedvalues)
# MSE(TV) =17.56
```

```python
lm_Tv = smf.ols(formula='Sales ~ TV ', data=df).fit()
lm_Tv.pvalues
# 3.32*e-41
```

```python
lm_Tv.rsquared
# R-squared(TV) = 0.6
```

```python
mean_squared_error(df['Sales'], lm_Tv.fittedvalues)
# MSE(TV) = 10.6
```

```python
lm_Newspaper = smf.ols(formula='Sales ~ Newspaper ', data=df).fit()
lm_Newspaper.pvalues
# 6.61*e-04
```

```python
lm_Newspaper.rsquared
# R-squared(Newspaper) = 0.06
```

```python
mean_squared_error(df['Sales'], lm_Newspaper.fittedvalues)
# MSE(Newspaper) = 25.30 
```

```python
# best MSE : MSE(TV)
# best R squared : R-squared(TV)
# best p values : pvalue (TV)
# best model is lm_Tv
```

```python
# normalisation
print(np.mean(df['Radio']/df['Sales']))
print(np.mean(df['TV']/df['Sales']))
# The ratios show that the 'TV' column contains high values comparing to the 'Radio' column 
```

```python
# In order to understand the effect of the regression coefficient of each variable on the variable 'Sales', 
# each vector should be set to one ( a colum of one)
# This way we will have : Sales = beta + intercept, therefore we will be able to see how each coefficient affect the variable 'SALES' 
```

```python
# If Tv = 0 ( No Tv advertisement ) , the sales will be equal to the intercept
# Thus, the intercept is the sales without any Tv advertisement
```

```python
# Multiple linear regression
lm_All = smf.ols(formula='Sales ~ Newspaper+TV + Radio', data=df).fit()
lm_All.summary()
# R-squared = 0.89
```

```python
lm_All.pvalues
# p value for the coefficient of regression of newspaper is so high 
# important predictors are TV , Radio
```

```python
mean_squared_error(df['Sales'], lm_All.fittedvalues)
```
