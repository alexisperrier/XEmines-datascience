```python
def format_arrdt(texte):
    texte = '751' + texte.split(' ')[1].replace('E','').replace('R','').zfill(2) 
    return int(texte)
```

```python
df['code_postal']  = df['ARRONDISSEMENT'].apply(lambda ardt : format_arrdt(ardt) )
```

```python
df.head()
```

```python
ardt = pd.read_csv("arrondissements.csv")
```

```python
ardt.head()
```

```python
arrds = ["PARIS 1ER ARRDT"]
for i in range(2,21):
    arrds.append("PARIS {}E ARRDT".format(i))

arrds
```

```python
arbres_par_arr = []
densite_km2 = []
for i in range(0,20):
    arbres_par_arr.append(len(df[df['ARRONDISSEMENT'] == arrds[i]].index))
    densite_km2.append(arbres_par_arr[i]/ardt.SUPERFICIE[i])
```

```python
density = np.vstack((arrds,densite_km2)).T
```

```python
density
```

Toutes les densités ont à peu près le même ordre de grandeurs.
On peut dire que l'arrondissement le plus vert est le 20ème suivi par le 13ème et le 7ème.
