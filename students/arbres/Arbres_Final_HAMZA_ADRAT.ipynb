#### Chercher Les arbres dans Paris (ARRONDISSEMENT contient "PARIS") :

```python
condition = df.ARRONDISSEMENT.str.contains('PARIS')
df[condition].ARRONDISSEMENT.value_counts()
```

```python
df = df[condition]
df.shape
```

```python
gb = df.groupby( by = 'ARRONDISSEMENT').count()
gb.reset_index(inplace=True)
gb
```

#### DataFrame contenant Arrondissement et Nbre_Arbre:

```python
gb.rename(columns={'DOMANIALITE': 'nbre_arbre'}, inplace = True)

gb = gb[['ARRONDISSEMENT', 'nbre_arbre']]

gb
```

#### Ajouter le code Postal:

```python
def format_arrdt (texte):
    '''
    PARIS 3E ARRDT => 75103
    '''
    texte = '751' + texte.split(' ')[1].replace('E','').replace('R','').zfill(2)
    return int(texte)

gb['code_postal'] = gb['ARRONDISSEMENT'].apply(lambda ardt : format_arrdt(ardt))
gb
```

```python
ardt = pd.read_csv('arrondissements-old.csv', sep=';')
ardt.head()
```

```python
ardt.rename(columns = {'C_ARINSEE': 'code_postal'}, inplace = True)
ardt.head()
```

#### Join !!

```python
gb = gb.merge(ardt, on='code_postal')
gb
```

#### Finalement ajouter la colonne Ratio (Nbre_Arbre/Surface):

```python
gb['ratio'] = gb['nbre_arbre']/gb['SURFACE']

gb.sort_values(by = 'ratio', ascending = False, inplace = True)

gb[['ARRONDISSEMENT', 'nbre_arbre', 'SURFACE', 'ration']]
```
