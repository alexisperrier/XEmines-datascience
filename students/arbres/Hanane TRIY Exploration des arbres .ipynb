**Des abérrations sont détectées, puisque l'hauteur d'un arbre ne peut pas dépasser 800kms! ainsi que pour les circonférences, donc on propose de supprimer toutes les lignes qui concernent par exemple des arbres à plus de 100 mètres de hauteur et 2 mètres comme circonférence en mettant des conditions sur ces deux variables : **

```python
df['HAUTEUR (m)'].value_counts()
```

```python
condition1 = (df['HAUTEUR (m)'] <= 100 ) & (df[ 'HAUTEUR (m)'] > 0)
df = df[condition1]
df['HAUTEUR (m)'].value_counts()
```
