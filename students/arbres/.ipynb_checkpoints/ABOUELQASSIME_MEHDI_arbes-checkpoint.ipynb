Nature des arbres par arrondissemen: Avant de commencer cette étape nous préferons tout d'abord de voir la corrélation entres les different variables. 

```python
import seaborn as sns
corr= data2.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

Nous remarquons qu'il n'y a pas une corrélation significante entre les variables, a part celle appelé "numéro" qui contient des 'NAN', ce qui ne pose pas de problème dans l'analyse des arbres par arrondissement, espèce ou dominalité.

Remarquable ou non ? 
Que signifie déja remarquable! 

```python
condition = data2['REMARQUABLE'] == 1
datarem =data2[condition]
condition2= data2['REMARQUABLE'] == 0
datanonrem=data2[condition2]
```

```python
condition3= (data2['REMARQUABLE'] != 1 ) & (data2['REMARQUABLE'] != 0)
datanan=data2[condition3]
```

```python
datarem['HAUTEUR (m)'].hist(bins = 30)

```

```python
datanonrem['HAUTEUR (m)'].hist(bins = 30)
```

```python
datanan['HAUTEUR (m)'].hist(bins = 30)
```

```python
Aprés avoir analysé les trois histogrammes nous remarquons que la hauteur a une influence evidente sur la catégorie remarquable, les arbres de hauteur plus que trois mètres sont ceux qui sont considéré remarquable, nous cocnluons que les observations en valeur "nan" sont peut etre mal rempli. 
```
