# Définition d'une nouvelle dataset contenant que les arrondissements dans Paris.

```python
data_paris['L_AR'] =data_paris['ARRONDISSEMENT'].str.replace('paris','')
data_paris
```

```python
data_paris = data_paris.groupby( by = 'ARRONDISSEMENT'  ).count()
data_paris.reset_index(inplace = True)
data_paris
```

```python
rania
```

C'est qu'à ce moment que j'ai constaté que la variable arrondissement et la variable L_AR ne sont pas de la méme forme, en absence du Paris et la présence de l'accent. Cette démarche n'a donc pas été bonne, on réessaie ultérieurement avec une autre qui va sûrement aboutir. 

# Définition de la fonction qui renvoie le code postal de chaque arrondissement

```python
def cd(arrond):
    arrond = '751' + arrond.split(' ')[1].replace('e','').replace('r','').replace('d','').zfill(2) 

    return int(arrond)
    

data_paris['code_postal']  = data_paris['ARRONDISSEMENT'].apply(lambda arrond : cd(arrond) )
data_paris
```

```python
arr
```

```python
arr.rename(columns = {'C_ARINSEE': 'code_postal'}, inplace = True)
arr.head()
```

```python
data_paris
```

# On combine les deux datasets en se basant sur la variable commune code postal

```python
data_paris = data_paris.merge(arr , on = 'code_postal')
data_paris
```

```python
data_paris.rename(columns={'DOMANIALITE': 'nbre_abres'}, inplace = True)

data_paris= data_paris[['ARRONDISSEMENT', 'nbre_abres','code_postal','SURFACE']]
data_paris
```

# Calcul du ratio des arbres par supérficie

```python
data_paris['ratio'] = data_paris['nbre_abres'] /  data_paris['SURFACE']

data_paris.sort_values(by = 'ratio' , ascending = False, inplace= True)
#data_paris[['ARRONDISSEMENT', 'nbre_abres', 'SURFACE', 'ratio']]
data_paris
```
